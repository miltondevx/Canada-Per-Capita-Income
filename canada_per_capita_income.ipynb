```python
import pandas as pd
import numpy as np
#from sklearn import linear_model
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:\Download\canada_per_capita_income.csv")
df.head()
```

```python
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year, df['per capita income (US$)'], color='red', marker='+')
```

```python
year = df.drop('per capita income (US$)',axis='columns')
#new_df = df.area
year.head()
```

```python
income = df['per capita income (US$)']
income.head()
```

```python
training = LinearRegression()
training.fit(year,income)
```

```python
# Predict for the year 2020
training.predict([[2020]])
```

```python
year_predict = 2020
income_predicted = training.predict([[2020]])
```
